````python
class Solution:
    def minMaxDifference(self, num: int) -> int:
        s = str(num)
        
        # T·∫°o max: ƒë·ªïi ch·ªØ s·ªë ƒë·∫ßu ti√™n kh√°c 9 th√†nh 9
        for c in s:
            if c != '9':
                max_val = int(s.replace(c, '9'))
                break
        else:
            max_val = num  # t·∫•t c·∫£ ƒë·ªÅu l√† 9
        
        # T·∫°o min: ƒë·ªïi ch·ªØ s·ªë ƒë·∫ßu ti√™n kh√°c 0 th√†nh 0
        for c in s:
            if c != '0':
                min_val = int(s.replace(c, '0'))
                break
        else:
            min_val = num  # t·∫•t c·∫£ ƒë·ªÅu l√† 0
        
        return max_val - min_val
````
