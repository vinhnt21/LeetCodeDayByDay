```python
from typing import Optional

# Định nghĩa một node trong cây
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

    # Hỗ trợ in cây theo kiểu duyệt trung thứ (in-order)
    def inorder(self):
        result = []
        def dfs(node):
            if not node:
                return
            dfs(node.left)
            result.append(node.val)
            dfs(node.right)
        dfs(self)
        return result

class Solution:
    def bstToGst(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        self.cumulative_sum = 0

        def reverse_inorder(node: Optional[TreeNode]):
            print(f"Đang xử lý node: {node.val if node else 'None'}")
            if not node:
                return
            
            reverse_inorder(node.right)

            print(f"Trước cập nhật: node.val = {node.val}, cumulative_sum = {self.cumulative_sum}")
            self.cumulative_sum += node.val
            node.val = self.cumulative_sum
            print(f"Sau cập nhật: node.val = {node.val}, cumulative_sum = {self.cumulative_sum}")
            
            reverse_inorder(node.left)

        reverse_inorder(root)
        return root

# --------------------------
# Example usage:

# Tạo cây ví dụ: [4,1,6,0,2,5,7,null,null,null,3,null,null,null,8]
# Ta sẽ tạo cây BST như vậy thủ công:
root = TreeNode(4,
        left=TreeNode(1,
            left=TreeNode(0),
            right=TreeNode(2,
                right=TreeNode(3))),
        right=TreeNode(6,
            left=TreeNode(5),
            right=TreeNode(7,
                right=TreeNode(8)))
    )

# In cây trước khi chuyển đổi
print("BST ban đầu (in-order):", root.inorder())

# Áp dụng chuyển đổi
sol = Solution()
new_root = sol.bstToGst(root)

# In cây sau khi chuyển đổi
print("Greater Tree (in-order):", new_root.inorder())
```
